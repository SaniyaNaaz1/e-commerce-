```python
import pandas as pd
import matplotlib.pyplot as plt
import glob
```

```python
product_file = glob.glob("output/day10_product_revenue/*.csv")[0]
df_product = pd.read_csv(product_file,header=None,names=["product","revenue"])
```

```python
category_file = glob.glob("output/day10_category_revenue/*.csv")[0]
df_category = pd.read_csv(category_file)
```

```python
print(df_product.head())
```

```python
print(df_product.columns)
```

```python
# Product Revenue Bar Chart
plt.figure(figsize=(8,5))
plt.bar(df_product['product'], df_product['revenue'])
plt.title("Revenue per Product")
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.tight_layout()
plt.savefig("output/product_revenue_chart.png")
plt.show()
```

```python
print(df_category.head())
```

```python
df_category = pd.read_csv(category_file, header=None)
df_category.columns = ['category', 'revenue']
```

```python
# Category Revenue Pie Chart
plt.figure(figsize=(6,6))
plt.pie(
    df_category['revenue'],
    labels=df_category['category'],
    autopct='%1.1f%%'
)
plt.title("Revenue by Category")
plt.savefig("output/category_revenue_chart.png")
plt.show()
```
